```python
from interarray.fileio import file2graph
from interarray.plotting import gplot, animate
from interarray.heuristics import CPEW, OBEW, ClassicEW
from interarray.interarraylib import calcload
from interarray.geometric import check_crossings
```

```python
vertices = file2graph("data/Gwynt y Mor.yaml")
```

```python
gplot(vertices, node_tag=None);
```

```python
G = CPEW(vertices, capacity=4)
calcload(G)
gplot(G, node_tag='label');
```

```python
G = OBEW(vertices, capacity=4)
calcload(G)
gplot(G, node_tag='load');
```

```python
animate(G)
```

```python
G = OBEW(vertices, capacity=4, rootlust='0.6*cur_capacity/capacity')
calcload(G)
gplot(G, node_tag='label');
```

```python
G = ClassicEW(vertices, capacity=4)
calcload(G)
gplot(G, node_tag='label');
```

```python
check_crossings(G);
```
